````python
#최솟값 찾기
from collections import deque   #deque 덱을 위한 라이브러리

N, L = map(int, input().split())    #N: 숫자 개수 L:최솟값을 구하는 범위

mydeque = deque()                   #데이터를 담을 덱 자료구조

now = list(map(int, input().split()))   #주어진 숫자데이터를 가지는 리스트

# mydeque[i][j]   i는 위치 값 인덱스 j는 범위 인덱스
for i in range(N):                      #숫자 개수만큼 반복
    while mydeque and mydeque[-1][0] > now[i]:          #덱의 마지막 위치에서부터 현재 값보다 큰 값은 덱에서 제거
        mydeque.pop()                                   #가장 오른쪽의 값을 삭제

        
    mydeque.append((now[i],i))                          #append(tuple(A[i], i)) 형태로??그리고 가장 오른쪽에 now 값 추가       인덱스도 같이 저장
    if mydeque[0][1] <= i - L:                          # 범위에서 벗어난 값은 덱에서 제거   #범위에서 벗어난 값
        mydeque.popleft()                               #범위에서 벗어난 값은 앞에값 뺴기
    print(mydeque[0][0], end=' ')
````

````python
#p1167_트리의 지름.py
#트리의 지름은 트리를 구성하는 노드 중 두 노드 사이의 거리가 가장 긴 것을 말한다.
from collections import deque

N = int(input())
A = [[] for _ in range(N + 1)]

for _ in range(N):
    Data = list(map(int, input().split()))
    index = 0
    S = Data[index]
    index += 1
    while True:
        E = Data[index]
        if E == -1:
            break
        V = Data[index + 1]
        A[S].append((E, V))
        index += 2

distance = [0] * (N + 1)            #리스트 초기화
visited = [False] * (N + 1)         #방문리스트 초기화

def BFS(v):                         #너비탐색
    queue = deque()                 #큐생성
    queue.append(v)                 #큐에 값 넣기
    visited[v] = True               #방문리스트 True
    while queue:                    #큐가 빌떄까지
        now_Node = queue.popleft()  #큐에 왼쪽깂 꺼내기(가장 먼저 넣은 값)
        for i in A[now_Node]:       #그 값에 연결된요소 for 문
            if not visited[i[0]]:   #방문리스트가 False라면
                visited[i[0]] = True    #True
                queue.append(i[0])      #값 넣고
                distance[i[0]] = distance[now_Node] + i[1]  #거리 리스트 업데이트

BFS(1)
Max = 1
for i in range(2, N + 1):
    if distance[Max] < distance[i]:
        Max = i

distance = [0] * (N + 1)
visited = [False] * (N + 1)
BFS(Max)
distance.sort()
print(distance[N])
````

````python
#원하는 정수 찾기
N = int(input())    #N: 수 개수 저장
A = list(map(int, input().split())) #수 데이터 리스트 저장
A.sort()                            #A 리스트 정렬
M = int(input())                    #M 탐색할 숫자 개수 저징
target_list = list(map(int, input().split()))           #탐색할 수 데이터 리스트 저장


for i in range(M):                  #M개수만큼 반복
    find = False                    
    target = target_list[i]         #찾아야하는 수
    # 이진탐색 시작
    start = 0
    end = len(A) - 1
    while start <= end:
        midi = int((start + end) / 2)
        midv = A[midi]
        if midv > target:
            end = midi - 1
        elif midv < target:
            start = midi + 1
        else:
            find = True
            break
    if find:
        print(1)
    else:
        print(0)
````

````python
#P2343_블루레이.py
#블루레이의 크기가 모두 같고 녹화 순서가 바뀌지 않아야 함 -> 이진 탐색 알고리즘
N, M = map(int, input().split())
A = list(map(int, input().split()))     #기타 레슨 데이터 저장 리스트
start = 0                               # 시작 인덱스
end = 0                                 # 종료 인덱스

for i in A:                         #리스트 탐색
    if start < i:
        start = i  # 레슨 최댓값을 시작 인덱스로 저장
    end += i  # 모든 레슨의 총합을 종료 인덱스로 저장
    
while start <= end:     #시작  인덱스 < = 종료 인덱스
    middle = int((start + end) / 2)     #중간인덱스
    sum = 0                             #레슨합
    count = 0                           #현재 사용한 블루레이 개수
    for i in range(N):  # 중간값으로 모든 레슨을 저장 할 수 있는지 확인
        if sum + A[i] > middle:
            count += 1
            sum = 0
        sum += A[i]
    if sum != 0:
        count += 1
    if count > M:
        start = middle + 1
    else:
        end = middle - 1
print(start)
````

````python
#P1744_수묶기
#주어진 수열의 각 수를 적절히 묶어 그 합을 최대로 만드는 프로그램을 작성하시오

from queue import PriorityQueue         #우선순위 큐 클래스 임포트

N = int(input())                        # 카드 묶음 개수
plusPq = PriorityQueue()                # 양수 우선순위 큐
mlusPq = PriorityQueue()                # 음수 우선순위 큐
one = 0                                 # 1의 개수 카운트
zero = 0                                # 0개의 개수 카운트

for i in range(N):                      # 4가지로 데이터 분리 저장
    data = int(input())
    if data > 1:
        plusPq.put(data * -1)           # 양수 내림차순 정렬을 위해 -1을 곱하여 저장
                                        # 가장 큰값부터 먼저 계산하고 싶은데
                                        # 우선순위 큐는 가장 작은값부터 정렬
                                        # -1를 곱해서 reverse 하는 느낌인거
    elif data == 1:                 #1일때
        one += 1
    elif data == 0:                 #0일때
        zero += 1
    else:                           #음수일때
        mlusPq.put(data)
sum = 0                             #전제 결과합 변수

while plusPq.qsize() > 1:           # 양수 처리 pluzPq.size()=1 일 때 종료
    first = plusPq.get() * -1       # 원소 꺼낼떄 -1 곱해줘야 함(넣을때 큰값부터 뽑고 싶어서 -1를 곱해서 넣음)
    second = plusPq.get() * -1      # 두번 째 값 꺼내기
    sum += first * second           # sum에다가 곱한거 더하기
    
if plusPq.qsize() > 0:              # plusPq.qzise ==1일 때 
    sum += plusPq.get() * -1        # sum에다가 -1 곱해서 더하기

while mlusPq.qsize() > 1:           # 음수 처리
    first = mlusPq.get()            # 음수 첫번째
    second = mlusPq.get()           # 음수 두번쨰
    sum += first * second           # 곱한거 sum에 더하기

if mlusPq.qsize() > 0:              #mlusPq.qsize()==1 하나 남았을 때
    if zero == 0:                   # 만약 0이 없으면
        sum += mlusPq.get()         # 그냥 sum에다가 더하기

sum += one                          # 1 처리(그냥 sum에다가 더하기)

print(sum)                          #결과값 출력
````

````python
#P1753_최단경로.py
import sys
input = sys.stdin.readline
from queue import PriorityQueue

V, E = map(int, input().split())
K = int(input())
distance = [sys.maxsize] * (V + 1)
visited = [False] * (V + 1)
myList = [[] for _ in range(V + 1)]
q = PriorityQueue()

for _ in range(E):
    u, v, w = map(int, input().split())  # 가중치가 있는 인접 리스트 저장
    myList[u].append((v, w))

q.put((0, K))  # K를 시작점으로 설정
distance[K] = 0
while q.qsize() > 0:
    current = q.get()
    c_v = current[1]
    if visited[c_v]:
        continue
    visited[c_v] = True
    for tmp in myList[c_v]:
        next = tmp[0]
        value = tmp[1]
        if distance[next] > distance[c_v] + value:  # 최소 거리로 업데이트
            distance[next] = distance[c_v] + value
            q.put((distance[next], next))
for i in range(1, V + 1):
    if visited[i]:
        print(distance[i])
    else:
        print("INF")
````

````python
#P1916_최소비용구하기
import sys
from queue import PriorityQueue
input = sys.stdin.readline

N = int(input())
M = int(input())
myList = [[] for _ in range(N + 1)]
dist = [sys.maxsize] * (N + 1)
visit = [False] * (N + 1)

for _ in range(M):
    start, end, weight = map(int, input().split())
    myList[start].append((end, weight))

start_idnex, end_index = map(int, input().split())

def dijkstra(start, end):
    pq = PriorityQueue()
    pq.put((0, start))
    dist[start] = 0
    while pq.qsize() > 0:
        nowNode = pq.get()
        now = nowNode[1]
        if not visit[now]:
            visit[now] = True
            for n in myList[now]:
                if not visit[n[0]] and dist[n[0]] > dist[now] + n[1]:
                    dist[n[0]] = dist[now] + n[1]
                    pq.put((dist[n[0]], n[0]))
    return dist[end]

print(dijkstra(start_idnex, end_index))
````

````python
#K번쨰촤단경로찾기.py
import sys
import heapq
input = sys.stdin.readline
N, M, K = map(int, input().split())
W = [[] for _ in range(N+1)]
distance = [[sys.maxsize] * K for _ in range(N+1)]

for _ in range(M):
    a, b, c = map(int, input().split())
    W[a].append((b,c))

pq = [(0,1)]
distance[1][0] = 0
while pq:
    cost,node = heapq.heappop(pq)
    for nNode, nCost in W[node]:
        sCost = cost + nCost
        if distance[nNode][K-1] > sCost:
            distance[nNode][K-1] = sCost
            distance[nNode].sort()
            heapq.heappush(pq, [sCost, nNode])

for i in range(1,N+1):
    if distance[i][K-1] == sys.maxsize:
        print(-1)
    else:
        print(distance[i][K-1])
````

````python
#정수를 1로 만들기
#X가 3으로 나누어떨어지면 3으로 나눈다.
#X가 2로 나누어 떨어지면 2로 나누다
#1을 뺀다
#정수N이 주어졌을 때 위와 같은 연산 3개를 적절히 사용해 1을 만들려고 한다. 연산을 사용하는
#횟수의 최솟값을 출력하시오.

import sys
input = sys.stdin.readline
N = int(input())            #구하고자 하는 수
D = [0]*(N+1)           #연산 횟수 최솟값을 저장하는 DP 테이블
D[1] = 0            #1일 때 연산 불필요
#D[i]= i 에서 1로 만드는데 걸리는 최소 연산횟수
for i in range(2, N+1):
    D[i] = D[i-1] + 1       #-1의 연산 표현
    if i % 2 == 0:
        D[i] = min(D[i], D[i//2] + 1)   #나누기 2연산
        #D[i//2] + 1 이 D[i]보다 작으면 변경
    if i % 3 == 0:
        D[i] = min(D[i], D[i//3] + 1)   #나누기 3연산
        #D[i//3] + 1 이 D[i]보다 작으면 변경
print(D[N])
````

````python
#퇴사 준비하기
#D[i] = i번째 날부터 퇴사날까지 벌 수 있는 최대 수입
#D[i]= D[i+1]       #오늘 시작되는 상담을 했을 때 퇴사일까지 끝나지 않는 경우
#D[i] = MAX(D[i+1], D[i+T[i]] + P[i]) #오늘 시작되는 상담을 했을 때 퇴사일 안에 끝나는 경우


import sys
input = sys.stdin.readline
N = int(input())    # 남은일수
D = [0]*(N+2)       # 오늘부터 퇴사일까지 벌 수 있는 최대 수입
T = [0]*(N+1)       # 상담에 필요한 일수
P = [0]*(N+1)       # 상담 완료 시 수입

for i in range(1, N+1):
    T[i], P[i] = map(int, input().split())      #상담에 필요한 일수 / 상담완료시 수입

for i in range(N, 0, -1):
    if i+T[i] > N+1:        #i번쨰 상담을 퇴사일까지 끝낼 수 없을 때
        D[i] = D[i+1]       #다음날로
    else:                                   #끝낼 수 있을 때
        D[i] = max(D[i+1], P[i] + D[i+T[i]])    #i번째 날의 최대 수입

print(D[1])
````

````python
###계단 수 구하기### - 소마대비문제
#ex)45656 이 수의 인접한 모든 자릿수의 차이는 1이다 이를 '계단 수'
#N이 주어질 때 길이가 N인 게단 수가 총 몇개 있는지 구하는 프로그램을 작성하시오.
#(0으로 시작하는 수는 없다)

#D[N][H]: 길이가 N인 계단에서 H 높이로 종료되는 계단 수를 만들 수 있는 경우의 수

#높이에 따른 점화식
# D[i][H] = D[i-1][H+1]           #H =0
# D[i][H] = D[i-1][H-1]           #H =9
# D[i][H] = D[i-1][H-1] + D[i-1][H+1] #H = 1-8

import sys
input = sys.stdin.readline
mod = 1000000000
N = int(input())                # 수 길이
D = [[0 for j in range(11)] for i in range(N+1)]

#D[N][H] :길이가 N일 때 높이 H로 끝나는 계단 수의 모든 경우의 순

for i in range(1, 10):
    D[1][i] = 1             #길이가 1일떄 만드는 높이 H로 끝나느
                            #단,0으로 시작하는 숫자는 없기 때문에 D[0][1]은 0으로 초기화
                            
for i in range(2, N+1):
    D[i][0] = D[i - 1][1]   #N에서 높이가 0이면 N-1에서는 높이가 1이어야 계단 수가 가능
    D[i][9] = D[i - 1][8]   #Nd에서 높이가 9이면 N-1에서는 높이가 8이어야 계단 수가 가능
    for j in range(1, 9):
        #높이가 1-8이면 N-1에서 자신보다 한 단계 위 또는 한 단계 아래 높이에서 올 수 있음
        D[i][j] = (D[i-1][j-1] + D[i-1][j+1]) % mod

sum = 0
for i in range(10):
    sum = (sum + D[N][i]) % mod     #정답값을 더할 때도 % 연산 수행

print(sum)


````

````python
#P13398_연속합2.py
#연속된 정수의 합 구하기
#연속된 몇개의 수를 선택해 구할 수 있는 합 중 가장 큰 합을 구ㅎ해보자
#또한 수열에서 수를 1개 제거할 수 있다.

#적절한 점화식 정의
# D[N]: 0에서 N까지 길이에서 N을 포함하여 연속으로 수를 선택하여 구할 수 있는 최대 합

#점화식의 의미
#L[N]: 왼쪽에서부터 N을 포함한 최대 연속 합을 나타냄. A[i]는 i자리의 수열의 수
#R[N]: 오른쪽에서부터 N을 포함한 최대 연속 합을 나타냄

#점화식 정의
#L[i] = Math.max(A[i], L[i-1]+A[i]);
#R[i] = Math.max(A[i], R[i+1] + A[i]);


N = int(input())                    #수열 크기
A = list(map(int, input().split())) #수열 데이터 저장 리스트

# 오른쪽 방향으로 index를 포함한 최대 연속합 구하기
L = [0] * N         #왼쪽에서 오른쪽으로 최대 연속합을 나타내는 테이블
L[0] = A[0]
result = L[0]
for i in range(1, N):
    L[i] = max(A[i], L[i - 1] + A[i])
    result = max(result, L[i])  # 제거하지 않았을 때는 기본 최댓값으로 저장

# 왼쪽 방향으로 index를 포함한 최대 연속 합 구하기
R = [0] * N          #오른쪽에서 왼쪽으로 최대 연속합을 나타내는 테이블
R[N - 1] = A[N - 1]
for i in range(N - 2, -1, -1):
    R[i] = max(A[i], R[i + 1] + A[i])

# L[i - 1] + R[i + 1] 2개의 구간 합 배열을 더하면 i번째 값을 제거한 효과를 얻음
for i in range(1, N - 1):
    temp = L[i - 1] + R[i + 1]
    result = max(result, temp)

print(result)
````

````python
##최장 공통 부분 수열 찾기###   #중요한 알고리즘이 담긴 문제##
###LCS : 최장 공통 부분 수열###

#대표적인 동적 계획법 문제###
#LCS 점화식을 이용해 값을 채웁니다.
#특정자리가 가리키는 행과 열의 문자열값을 비교해 값이 같으면 대각선 왼쪽위의 값에 1을 더한값을 저장
# DP[i][j] = DP[i - 1][j - 1] + 1  # 같은 문자열일 때 왼쪽 대각선 값 + 1

#비교한 값이 다르면 테이블의 왼쪽과 위쪽 값 중 큰 값을 선택해 저장합니다.
# DP[i][j] = max(DP[i - 1][j], DP[i][j - 1])  # 다르면 왼쪽과 위의 값 중 큰 수

import sys
sys.setrecursionlimit(10000)
input = sys.stdin.readline
A = list(input())           #첫번쨰 문자열
A.pop()  # \n 문자열 제거
B = list(input())           #두번쨰 문자열
B.pop()  # \n 문자열 제거
DP = [[0 for j in range(len(B) + 1)] for i in range(len(A) + 1)]    #2차원 점화식 테이블
Path = []       #LCS 저장리스트

#점화식 테이블에 값 넣긴
for i in range(1, len(A) + 1):
    for j in range(1, len(B) + 1):
        if A[i - 1] == B[j - 1]:
            DP[i][j] = DP[i - 1][j - 1] + 1  # 같은 문자열일 때 왼쪽 대각선 값 + 1
        else:
            DP[i][j] = max(DP[i - 1][j], DP[i][j - 1])  # 다르면 왼쪽과 위의 값 중 큰 수

print(DP[len(A)][len(B)])      #대각선 가장 끝의 값


# LCS 구현 함수
def getText(r, c):              #재귀 형태로 구현
    if r == 0 or c == 0:
        return
    if A[r - 1] == B[c - 1]:  # 같으면 LCS에 기록하고 왼쪽위로 이동
        Path.append(A[r - 1])
        getText(r - 1, c - 1)
    else:  # 다르면 왼쪽과 중 큰 수로 이동
        if DP[r - 1][c] > DP[r][c - 1]:
            getText(r - 1, c)
        else:
            getText(r, c - 1)


getText(len(A), len(B))

for i in range(len(Path) - 1, -1, -1):
    print(Path.pop(i), end='')
print()

````

````python
#가장 큰 정사각형 찾기
#크기가 n*m이고 0과 1로만 이루어진 배열이 있다. 이 배열에서 1로 된 가장 큰 정사각형의 크기를 구하는 프로그램을 작성하시오.
#정사각형긔 넓이 ~ 큰 정사각형의 한변의 길이

#점화식 정의
# D[i][j] = i,j의 위치를 오른쪽 아래 꼭짓점으로 정하고, 해당 자리에서 그릴 수 있는가장 큰 정사각형으ㅢ 길이

# D[i][j] = MIN(D[i-1][j-1], D[i][j-1], D[i-1][j]) +1 #현재 자리의 원래 값이 1
# D[i][j] = 0   # 현재 자리의 원 래 값이 0

import sys
input = sys.stdin.readline

D = [[0 for j in range(1001)] for i in range(1001)]     #i,j 위치에서 왼쪽 위로 만들 수 있는 최대 정사각형 길이를 저장하는 리스트
N, M = map(int, input().split())                        # N(리스트의 세로길이) M(리스트의 가로 길이)
                        

max = 0                 #최댓값 저장

for i in range(0, N):           #N만큼 반복
    numbers = list(input())     #데이터를 한줄씩 저장
    for j in range(0, M):       #M만큼 반복
        D[i][j] = int(numbers[j])   #D[i][j]에  number[j] 값을 저장
        if D[i][j] == 1 and j > 0 and i > 0:            #D[i][j]값이 1이면
            D[i][j] = min(D[i-1][j-1], D[i-1][j], D[i][j-1]) + D[i][j]      #자신의 위, 왼족, 대가선 값 중 최솟값 +1 값을 저장
        if max < D[i][j]:                                                   #D[i][j]의 값이 최대값보다 크면:
            max = D[i][j]                                                   #최대값 업데이트

print(max*max)          #결과값 출력
````

````python
#빌딩 순서 구하기
# 가장 왼쪽에서 봤을 때 보이는 빌딩의 수 L,가장 오른쪽에서 봤을 때 보이는 빌딩의 수 R이 주어졌을 때 가능한
# 빌딩의 순서의 경우의 수를 구하는 프로그램을 작성

#점화식 정의
# D[N][L][R] : N개의 빌딩이 있고 왼쪽에서 L개, 오른쪽에서 R개가 보일 때 가능할 경우의 수

#N-1개의 빌딩에서 왼쪽에 작은 빌딩을 추가할 때
# D[N-1][L-1][R]

#N-1개의 빌딩에서 오른쪽에 작은 빌딩을 추가할때
# D[N-1][L-1][R]

#가운데는 
# D[N-1][L][R]*(N-2)            N-2개의 위치에 배치할 수 있으므로  N-2를 곱한다.

# D[N][L][R] = D[N-1][L-1][R] +  D[N-1][L-1][R] + D[N-1][L][R]*(N-2)
# D[1][1][1]=1

import sys
input = sys.stdin.readline
mod = 1000000007
N, L, R = map(int, input().split())  #N개의 빌딩이 있고 왼쪽에서 L개, 오른쪽에서 R개
D = [[[0 for k in range(101)] for j in range(101)] for i in range(101)]
D[1][1][1] = 1  # 빌딩이 1개이면 놓을 수 있는 경우의 수는 1개

for i in range(2, N + 1):       #i from N
    for j in range(1, L + 1):   #j from L
        for k in range(1, R + 1):   #k from R
            D[i][j][k] = (D[i-1][j][k] * (i-2) + (D[i-1][j][k-1] + D[i-1][j-1][k])) % mod

print(D[N][L][R])
````

````python
#DDR
#점화식 정의
# D[N][L][R] N개의 수열을 수행한 후 왼발의 위치가 L 오른발의 위치가 R일 때 최소 누적힘
# D[N][L][R]  = MIN(D[N-1][L][i] + mp[i][R])

import sys
input = sys.stdin.readline
# dp[N][L][R] = N개 수열을 수행했고, 왼쪽이 L 오른쪽이 R에 있을 때 최소 누적
# 충분히 큰 수로 초기화
dp = [[[sys.maxsize for k in range(5)] for j in range(5)] for i in range(100001)]
# 한 발을 이동할 때 드는 힘을 미리 저장하기(mp[1][2] → 1에서 2로 이동할 때 드는 힘)
mp = [[0, 2, 2, 2, 2],
      [2, 1, 3, 4, 3],
      [2, 3, 1, 3, 4],
      [2, 4, 3, 1, 3],
      [ 2, 3, 4, 3, 1]]

s = 1
dp[0][0][0] = 0

inputList =  list(map(int, input().split()))
index = 0

while inputList[index] != 0:
    n = inputList[index]
    for i in range(5):
        if n == i:   #두 발이 같은 자리에 있을 수 없음
            continue
        for j in range(5):
            # 오른발을 옮겨 현재 모습이 됐을 때 최소 힘 저장
            dp[s][i][n] = min(dp[s - 1][i][j] + mp[j][n], dp[s][i][n])

    for j in range(5):
        if n == j:   #두 발이 같은 자리에 있을 수 없음
            continue
        for i in range(5):
            # 왼발을 옮겨 현재 모습이 됐을 때 최소 힘 저장
            dp[s][n][j] = min(dp[s - 1][i][j] + mp[i][n], dp[s][n][j])
    s += 1
    index += 1
s -= 1  # 마지막 이동 횟수로 index 조정
result = sys.maxsize
for i in range(5):
    for j in range(5):
        result = min(result, dp[s][i][j])   # 모두 수행한 후 최솟값 찾기
print(result)   # 최솟값 출력
````

````python
# 행렬 곱 연산 횟수의 최솟값 구하기
# 점화식 정의
# D[i][j] : i-j 구간의 행렬을 합치는 데 드는 최소 연산 횟수


import sys

input = sys.stdin.readline
N = int(input())
M = []
D = [[-1 for j in range(N + 1)] for i in range(N + 1)]

M.append((0, 0))
for i in range(N):
    x, y = map(int, input().split())
    M.append((x, y))

def excute(s, e):
    result = sys.maxsize
    if D[s][e] != -1:
        return D[s][e]
    if s == e:
        return 0
    if s + 1 == e:
        return M[s][0] * M[s][1] * M[e][1]
    for i in range(s, e):
        result = min(result, M[s][0] * M[i][1] * M[e][1] + excute(s, i) + excute(i + 1, e))
    D[s][e] = result
    return D[s][e]

print(excute(1, N))
````

````python
#외판원의 순회 경로 짜기 TSP 문제
#N과 비용 행렬이 주어졌을 때 가장 적은 비용을 들이는 외판원의 순회 여행 경로를 구하는 프로그램을 작성하시오

#점화식 정의

import sys
input = sys.stdin.readline

N = int(input())        #도시 개수
W = []                  #비용 저장 리스트

for i in range(N):
 W.append([])
 W[i] = list(map(int, input().split()))
 
D = [[0 for j in range(1 << 16)] for i in range(16)]

def tsp(c,v):               #완전탐색의 형태로 수행
    if v == (1<<N)-1:       #모든 노드를 방문한 경우
        if W[c][0] == 0:
            return float('inf')
        else:
            return W[c][0]
    if D[c][v] != 0:        #이미 방문한 노드인 경우 바로 리턴
        return D[c][v]
    min_val = float('inf')
    for i in range(0, N):
        #방문한 적이 없고, 갈 수 있는 도시인 경우
        if (v & (1<<i)) == 0 and W[c][i] != 0:
            min_val = min(min_val, tsp(i, (v | (1 << i))) + W[c][i])
    D[c][v] = min_val
    return D[c][v]

print(tsp(0,1))
````

````python
#가장 길게 증가하는 부분 수열 찾기

import sys
input = sys.stdin.readline
N = int(input())                    #수열 개수
A = list(map(int, input().split())) #수열 데이터 저장리스트
A.insert(0, 0)

index = 0
maxLength = 1
B = [0] * 1000001
D = [0] * 1000001
ans = [0] * 1000001
B[maxLength] = A[1]
D[1] = 1

# 바이너리 서치 구현
def binarysearch(l, r, now):            #현재 수열이 들어갈 수 있는 위치를 빠르게 찾기 위한 함수
    while l < r:
        mid = (l + r) // 2
        if B[mid] < now:
            l = mid + 1
        else:
            r = mid
    return l

for i in range(2, N + 1):
    if B[maxLength] < A[i]:  # 가장 마지막 수열보다 현재 수열이 큰 경우
        maxLength += 1
        B[maxLength] = A[i]
        D[i] = maxLength
    else:  # 바이너리 서치를 이용해 현재 수열이 들어갈 index 찾기
        index = binarysearch(1, maxLength, A[i])
        B[index] = A[i]
        D[i] = index

print(maxLength)
index = maxLength
x = B[maxLength] + 1
for i in range(N, 0, -1):
    if D[i] == index and A[i] < x:
        ans[index] = A[i]
        x = A[i]
        index -= 1

for i in range(1, maxLength + 1):
    print(ans[i], end=' ')

````
